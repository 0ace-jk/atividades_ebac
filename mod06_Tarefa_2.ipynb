# Módulo 06 Tarefa 2

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(x=df['IDADEMAE'])
plt.title('Idade da mãe')
plt.xlabel('Idade')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(x=df['PESO'])
plt.title('Peso do bebê')
plt.xlabel('Peso em gramas')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.histplot(x=df['IDADEMAE'], binwidth=1)
plt.title('Idade da mãe')
plt.xlabel('Idade')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.histplot(x=df['PESO'], binwidth=100)
plt.title('Peso do bebê')
plt.xlabel('Peso em gramas')
plt.show()
```
