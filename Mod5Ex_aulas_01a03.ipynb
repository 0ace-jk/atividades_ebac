```python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 2) seu código aqui

for col in sinasc.columns:
    print(col, sinasc[col].isna().sum())
```

```python
# 3) seu código aqui

sinasc_red = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO',
 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']].copy()
print('A quantidade de valores faltantes é dada na seguinte tabela:')
sinasc_red.isna().sum()
```

```python
# 4) seu código aqui

sinasc_red.dropna(subset=['APGAR5'], inplace=True)
print(sinasc_red.isna().sum())
print('\n')
print(f'O numero de linhas agora é de: {sinasc_red.shape[0]}')
```

```python
# 5) seu código aqui

sinasc_red['ESTCIVMAE'] = sinasc_red['ESTCIVMAE'].fillna(9)
sinasc_red['CONSULTAS'] = sinasc_red['CONSULTAS'].fillna(9)
sinasc_red.isna().sum()
```

```python
# 6) Seu código aqui

sinasc_red['QTDFILVIVO'] = sinasc_red['QTDFILVIVO'].fillna(0)
sinasc_red.isna().sum()
```

```python
# 7) seu código aqui

sinasc_red['ESCMAE'] = sinasc_red['ESCMAE'].fillna(9)

sinasc_red['GESTACAO'] = sinasc_red['GESTACAO'].fillna(9)

sinasc_red['GRAVIDEZ'] = sinasc_red['GRAVIDEZ'].fillna(9)

# Decido inserir o valor '9' (Ignorado) para preencher esse valores faltantes
# pois não pretendo compremeter nenhum estudo futuro desses valores
# caso seja necessario, deveremos utilizar algum metodo para todos os valores '9'

sinasc_red.isna().sum()
```

```python
# 8) seu código aqui
sinasc_red['APGARCAT'] = sinasc_red['APGAR5'].apply(lambda x: 'normal' if x >= 8 else ('asfixia leve' if x >= 6 else ('asfixia moderada' if x >= 4 else 'asfixia severa')))
sinasc_red['APGARCAT'].value_counts()
```

```python
# 9) seu código aqui

sinasc_red.rename(columns={'LOCNASC': 'loc_nasc',
                           'IDADEMAE': 'idade_mae',
                           'ESTCIVMAE': 'est_civ_mae',
                           'ESCMAE': 'esc_mae',
                           'QTDFILVIVO': 'qtd_fil_vivo',
                           'GESTACAO': 'gestacao',
                           'GRAVIDEZ': 'gravidez',
                           'CONSULTAS': 'consultas',
                           'APGAR5': 'apgar_5',
                           'APGARCAT': 'apgar_cat',
                           })
sinasc_red.head()
```
