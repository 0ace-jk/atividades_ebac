# Random forest vs AdaBoost:

## 5 diferenças entre Random Forest e AdaBoost:

Diferença | Random Forest | AdaBoost
--- | --- | ---
Tipo de Algoritmo         | Random Forest é um algoritmo de ensemble baseado em bagging                                     | AdaBoost é um algoritmo de ensemble baseado em boosting.
Construção de Modelos     | Random Forest constrói múltiplas árvores de decisão em paralelo                                 | AdaBoost constrói árvores sequencialmente, cada uma corrigindo os erros da anterior.
Peso dos Modelos          | No Random Forest, todas as árvores têm o mesmo peso na decisão final                            | No AdaBoost, árvores com melhor desempenho têm mais peso.
Robustez a Ruídos         | Random Forest é mais robusto a ruídos e outliers devido à média das previsões de várias árvores | AdaBoost pode ser mais sensível a outliers.
Velocidade de Treinamento | Random Forest geralmente treina mais rápido, pois as árvores são construídas em paralelo        | AdaBoost pode ser mais lento devido à sua natureza sequencial.


# Exemplo de uso do AdaBoost com validação cruzada


```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings("ignore")

X, y = load_iris(return_X_y=True)
clf = AdaBoostClassifier(n_estimators=50)
scores = cross_val_score(clf, X, y, cv=5)
display(scores.mean())

import matplotlib.pyplot as plt

plt.plot(scores)
plt.title("Acurácia do AdaBoost")
plt.xlabel("Fold")
plt.ylabel("Acurácia")
plt.show()
```

# 5 Hyperparâmetros importantes do AdaBoost:
- n_estimators: Número de estimadores fracos (árvores de decisão) a serem usados no ensemble.
- learning_rate: Taxa de aprendizado que controla o peso de cada estimador fraco.
- estimator: O estimador fraco a ser usado (por padrão, é uma árvore de decisão de profundidade 1).
- algorithm: O algoritmo de boosting a ser usado (por exemplo, 'SAMME' ou 'SAMME.R').
- random_state: Semente para o gerador de números aleatórios, garantindo reprodutibilidade dos resultados.

```python
clf = AdaBoostClassifier(
    n_estimators=150, 
    learning_rate=0.1, 
    algorithm='SAMME', 
    random_state=42, 
    estimator=None
)
scores = cross_val_score(clf, X, y, cv=5)
display(scores.mean())

plt.plot(scores)
plt.title("Acurácia do AdaBoost com hiperparâmetros ajustados")
plt.xlabel("Fold")
plt.ylabel("Acurácia")
plt.show()
```

# Utilizando o GridSearchCV para encontrar os melhores hiperparâmetros do AdaBoost

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 150, 200, 500],
    'learning_rate': [0.005, 0.01, 0.1, 1]
}
grid_search = GridSearchCV(clf, param_grid, cv=5)
grid_search.fit(X, y)
display(grid_search.best_params_)
display(grid_search.best_score_)
```
