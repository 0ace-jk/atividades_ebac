### Fórmula de Bhaskara

A solução para uma função de segundo grau depende das raízes (valores de x). Como já vimos, os coeficientes precisam ser números reais e o angular diferente de zero. Sendo assim, temos a seguinte fórmula de Bhaskara:<br><br>

## $$X = \frac{-b \pm \sqrt{b^2 - 4.a.c}}{2.a}$$
<center><em>Fórmula de Bhaskara.</em></center><br>

Para melhor compreensão o seu cálculo é dividido em duas partes: discriminante da equação e operações para determinar as raízes.<br>

#### **Discriminante**
A figura dentro da raiz na fórmula de Bhaskara é nomeada de discriminante. Seu símbolo é a letra grega delta e apresenta a determinada fórmula:<br><br>

## $$\Delta = b^2 - 4.a.c$$
<center><em>Fórmula do discriminante.</em></center><br>

- Se o delta for maior que zero, a equação terá dois valores reais e distintos.
- Se o delta for igual a zero, a equação terá somente um valor real ou dois resultados iguais.
- Se o delta for menor que zero, a equação não possuirá valores reais.
