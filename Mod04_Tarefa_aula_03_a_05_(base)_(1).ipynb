## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

# Acertos Matematica
percentual_mat = df_mat['Qt_acertos'].sum() / len(df_mat) / 24
print(f'Os alunos acertam {(percentual_mat * 100).round(2)}% das questões de Matematica')
# Acertos Portugues
percentual_por = df_por['Qt_acertos'].sum() / len(df_por) / 18
print(f'Os alunos acertam {(percentual_por * 100).round(2)}% das questões de Portugues')
# Acertos Geografia
percentual_geo = df_geo['Qt_acertos'].sum() / len(df_geo) / 8
print(f'Os alunos acertam {(percentual_geo * 100).round(2)}% das questões de Geografia')
# Acertos Ingles
percentual_ing = df_ing['Qt_acertos'].sum() / len(df_ing) / 8
print(f'Os alunos acertam {(percentual_ing * 100).round(2)}% das questões de Ingles')
# Acertos Historia
percentual_his = df_his['Qt_acertos'].sum() / len(df_his) / 8
print(f'Os alunos acertam {(percentual_his * 100).round(2)}% das questões de Historia')
# Acertos Fisica
percentual_fis = df_fis['Qt_acertos'].sum() / len(df_fis) / 12
print(f'Os alunos acertam {(percentual_fis * 100).round(2)}% das questões de Fisica')
# Acertos Quimica
percentual_qui = df_qui['Qt_acertos'].sum() / len(df_qui) / 12
print(f'Os alunos acertam {(percentual_qui * 100).round(2)}% das questões de Quimica')
````

````python
# aqui eu crio Data Frames específicos para cada matéria
df_mat_alunos = df_mat.rename(columns={'Qt_acertos': 'Qt_acertos_mat'})
df_por_alunos = df_por.rename(columns={'Qt_acertos': 'Qt_acertos_por'})
df_geo_alunos = df_geo.rename(columns={'Qt_acertos': 'Qt_acertos_geo'})
df_ing_alunos = df_ing.rename(columns={'Qt_acertos': 'Qt_acertos_ing'})
df_his_alunos = df_his.rename(columns={'Qt_acertos': 'Qt_acertos_his'})
df_fis_alunos = df_fis.rename(columns={'Qt_acertos': 'Qt_acertos_fis'})
df_qui_alunos = df_qui.rename(columns={'Qt_acertos': 'Qt_acertos_qui'})
df_alunos = pd.DataFrame(np.arange(1, 101),columns = ['id_aluno'])
````

````python
# neste codigo eu concateno os dataframes das materias para um unico dataframe
# e ja indexo a coluna id_alunos
alunos = [df_alunos,
          df_mat_alunos,
          df_por_alunos,
          df_geo_alunos,
          df_ing_alunos,
          df_his_alunos,
          df_fis_alunos,
          df_qui_alunos]
df_resultado = pd.concat(alunos, axis=1)
df_resultado.set_index('id_aluno', inplace = True)
````

````python
# 2) Seu código aqui

df_resultado['Qt_acertos_total'] = df_resultado.loc[:,'Qt_acertos_mat':].sum(axis=1)
df_resultado['Qt_acertos_total']
````

````python
# 3) Seu código aqui

df_resultado['percentual_acertos'] = df_resultado['Qt_acertos_total'] / 90
df_resultado['percentual_acertos']
````

````python
# 4) Seu código aqui

df_resultado['aluno_aprovado'] = df_resultado['percentual_acertos'] > 0.45
print(f"A quantidade de alunos aprovados foi de {df_resultado['aluno_aprovado'].sum()} alunos")
````
