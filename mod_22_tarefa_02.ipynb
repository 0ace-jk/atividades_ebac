#### 1. Carregue os dados da tarefa anterior

````python
import pandas as pd
import plotly.express as px
from plotly import graph_objects
from plotly.graph_objs import Layout
````

````python

data = pd.read_csv('dados_yahoo.csv', header=[0, 1], index_col=0)

data.index = pd.to_datetime(data.index)

data.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
adj_close = data['Adj Close']

media_movel_90d = adj_close.rolling(window=90).mean()

desvio_padrao_90d = adj_close.rolling(window=90).std()

fig_media_movel = px.line(media_movel_90d,
                          title='Média Móvel de 90 dias dos Ativos',
                          labels={'value': 'Preço Médio', 'Date': 'Data', 'variable': 'Ativo'},
                          template='plotly_dark')
fig_media_movel.show()

fig_desvio_padrao = px.line(desvio_padrao_90d,
                            title='Desvio Padrão Móvel de 90 dias (Volatilidade)',
                            labels={'value': 'Desvio Padrão', 'Date': 'Data', 'variable': 'Ativo'},
                            template='plotly_dark')
fig_desvio_padrao.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['ABEV3.SA']*len(lista_indicadores)

colunas = list(zip(lista_indicadores, ativo))

ambev = data.loc['2025-08-31':, colunas]
ambev.columns = lista_indicadores

graph = {
    'x' : ambev.index,
    'open' : ambev.Open,
    'close' : ambev.Close,
    'high' : ambev.High,
    'low' : ambev.Low,
    'type' : 'candlestick',
    'name' : 'ABEV3.SA',
    'showlegend' : False
}

layout = Layout(
    paper_bgcolor='rgba(225,225,225,1)',
    plot_bgcolor='rgba(225,225,225,1)'
)

grafico = graph_objects.Figure(
    data=[graph],
    layout_title='Gráfico Candlestick - ABEV3.SA',
    layout=layout
)

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='DarkGray')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='DarkGray')
````
