    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.

```python
# Carregando bilbliotecas
# Utilizando yfinance ao invés de pandas_datareader

# from pandas_datareader import data as web
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.express as px
from plotly import graph_objects
```

```python
# Baixando dados do Yahoo Finance e salvando em CSV

ativos = None
ativos = ['ITUB4.SA', 'OIBR3.SA', 'ABEV3.SA']
dt_ini, dt_fim = '2019-01-01', '2021-12-31'
data = yf.download(ativos, start=dt_ini, auto_adjust=False)
data.head()
data.to_csv('dados_yahoo.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Carregando dados do CSV

df = pd.read_csv('dados_yahoo.csv', header=[0,1], index_col=0, parse_dates=True)
```

```python
# Calculando a amplitude diária das ações com um loop for

for ativo in ativos:
    df[('Amplitude', ativo)] = df[('High', ativo)] - df[('Low', ativo)]
df.head()
```

```python
# Reorganizando o DataFrame para formato vertical

fechamento_vertical = df['Amplitude'].stack().reset_index()
fechamento_vertical.columns = ['Data', 'Ticker', 'Amplitude']
fechamento_vertical
```

```python
# Plotando a amplitude diária das ações

fig = px.line(fechamento_vertical, x='Data', y='Amplitude', title='Amplitude Diária das Ações', color='Ticker', template='plotly_dark')
fig.show()
```

Como podemos observar, a volatilidade da 'OIBR3.SA' é maior que a dos outros dois ativos no período analisado, porem ao observar os dados mais recentes (2025+) podemos notar que a 'ITUB4.SA' apresenta uma volatilidade maior.
