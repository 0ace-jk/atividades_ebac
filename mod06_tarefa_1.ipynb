```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Módulo 6 tarefa 1

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
df = pd.read_csv('/content/SINASC_RO_2019.csv')
df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
sns.countplot(x='APGAR1', data=df, ax=ax1)
plt.xlabel('')
plt.ylabel('')
plt.title('APGAR 1')
ax2 = fig.add_subplot(1, 2, 2)
sns.countplot(x='APGAR5', data=df, ax=ax2)
plt.xlabel('')
plt.ylabel('')
plt.title('APGAR 5')
plt.subplots_adjust(wspace=.5)

ax1.tick_params(axis = 'x', rotation=90)
ax2.tick_params(axis = 'x', rotation=90)
plt.show()
```

```python
ax = sns.countplot(x='SEXO', data=df)
ax.bar_label(ax.containers[0])
plt.xlabel('')
plt.ylabel('')
plt.title('Sexo')
```

```python
ax = sns.countplot(x='GESTACAO', data=df)
plt.xticks(rotation=45)
ax.bar_label(ax.containers[0])
plt.xlabel('')
plt.ylabel('')
plt.title('Semanas de gestação')
```

```python
ax = sns.countplot(x='ESTCIVMAE', data=df)
ax.bar_label(ax.containers[0])
plt.xlabel('')
plt.ylabel('')
plt.title('Estado civil')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar5_medio = df.groupby('CONSULTAS')['APGAR5'].mean()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=apgar5_medio.index, y=apgar5_medio.values)
ax.bar_label(ax.containers[0])
plt.xlabel('Número de Consultas Pré-natal')
plt.ylabel('Média APGAR5')
plt.title('Média de APGAR5 por Número de Consultas Pré-natal')
plt.xticks(rotation=45)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
plt.subplots_adjust(hspace=.5)

ax1 = sns.countplot(x='SEXO', data=df, ax=axes[0,0])
ax1.set_xlabel('')
ax1.set_ylabel('')
ax1.set_title('Sexo')
ax1.bar_label(ax1.containers[0])
ax1.tick_params(axis = 'x', rotation=45)

ax2 = sns.countplot(x='GESTACAO', data=df, ax=axes[0,1])
ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_title('Semanas de gestação')
ax2.bar_label(ax2.containers[0])
ax2.tick_params(axis = 'x', rotation=45)

ax3 = sns.countplot(x='ESTCIVMAE', data=df, ax=axes[1,0])
ax3.set_xlabel('')
ax3.set_ylabel('')
ax3.set_title('Estado civil')
ax3.bar_label(ax3.containers[0])

ax4 = sns.countplot(x='CONSULTAS', data=df, ax=axes[1,1])
ax4.set_xlabel('')
ax4.set_ylabel('')
ax4.set_title('Número de Consultas Pré-natal')
ax4.bar_label(ax4.containers[0])
plt.show()
```
