# Cálculo - Tarefa 02

1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Definir a função
def f(x):
    return x**3

# 2. Definir os limites da integração e o número de retângulos
a = 1  # Limite inferior
b = 2  # Limite superior
n = 50 # Número de retângulos

# 3. Calcular a largura de cada retângulo
largura_retangulo = (b - a) / n

# 4. Calcular os pontos médios de cada retângulo para determinar sua altura
pontos_medios = np.linspace(a + largura_retangulo/2, b - largura_retangulo/2, n)

# 5. Calcular a altura de cada retângulo
alturas = f(pontos_medios)

# 6. Somar a área de todos os retângulos (altura * largura)
area_total_numerica = np.sum(alturas * largura_retangulo)

print(f"Número de retângulos usados: {n}")
print(f"A área calculada numericamente é: {area_total_numerica:.4f}")

# Visualização Gráfica
x_curva = np.linspace(a, b, 100)
y_curva = f(x_curva)

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(pontos_medios, alturas, width=largura_retangulo, edgecolor='gray', alpha=0.6, label=f'{n} Retângulos')

ax.plot(x_curva, y_curva, 'b-', linewidth=2, label='$f(x) = x^3$')

ax.set_title('Integração Numérica (Método dos Retângulos)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

<center><h3> Vamos encontrar a area sob a curva de forma analítica:</h3></center>

<center>1. Encontrar a antiderivada de 

$$
x^3
$$

A regra de integração para potências diz que:

$$
x^n dx = \frac{x^{n+1}}{n+1}
$$ 

Então:

$$
F(x) = \frac{x^3+1}{3+1} = \frac{x^{4}}{4}
$$
</center>
<center>2. Avaliar a antiderivada nos limites de integração para os pontos 2 e 1:

$$
F(2) - F(1) = \left[ \frac{x^4}{4} \right]_{1}^{2} = \frac{2^4}{4} - \frac{1^4}{4}
$$

</center>
<center>3. Calculando o resultado final:

$$
\frac{2^4}{4} - \frac{1^4}{4} = \frac{16}{4} - \frac{1}{4} = \frac{15}{4} = 3.75
$$

</center>
